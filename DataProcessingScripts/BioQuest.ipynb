```python
df = pd.read_csv("../Data/0.2.0-BioQuest.csv")
df = df.rename(columns={"CompoundName" : "Compound", "BoilingPoint(C)" : "BoilingPoint/C", "MeltingPoint(C)" : "MeltingPoint/C"}) #Renaming columns

boilingPoints = df["BoilingPoint/C"].values #Extracting points as lists
meltingPoints = df["MeltingPoint/C"].values

def pointCleaner(pointList):
    """ 
    This function cleans the points by removing any special characters and converting them to float values.
    
    Args:
    pointList : list of points to be cleaned
    
    Returns:
    cleanedPoints : list of cleaned points"""

    cleanedPoints = []; nErrors = [] #Initialising lists
    for point in pointList: #Iterating through the list
        if point is not np.nan:

            point = point.replace(">", "") #Removing special characters
            point = point.strip()

            if "d" in point: #Handling points with "d" in them
                point = point.split("d")[0]

            try:
                spltChar = "-" #Splitting at "-" if present
                spltPoints = point.split(spltChar)
                val = spltChar.join(spltPoints[:1]) #Joining the split points to get 1st and second value of the list without removing negative sign (splits on second "-" if present)
                val2 = spltChar.join(spltPoints[1:])
                cleanedPoints.append(np.mean([float(val), float(val2)])) #Calculating mean of the two values
            except:
                try:
                    cleanedPoints.append(float(point)) #Converting to float if only one value is present
                except:
                    nErrors.append(point) #Appending to error list if error occurs
                    cleanedPoints.append(np.nan)
        else:
            cleanedPoints.append(np.nan)

    print(nErrors) #Printing errors
    print(len(nErrors))
    return cleanedPoints

df["BoilingPoint/C"] = pointCleaner(boilingPoints) #Cleaning points
df["MeltingPoint/C"] = pointCleaner(meltingPoints)

df.to_csv("../Data/0.3.1-BioQuest.csv", index=False) #Saving cleaned data
```
