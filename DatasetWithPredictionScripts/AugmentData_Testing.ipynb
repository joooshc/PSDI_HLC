```python
import pandas as pd
import numpy as np
import sys
sys.path.append('../')
from MachineLearning.MyFuncs.checkDirs import *
```

```python
og = pd.read_csv("../Data/Combined/LargeFiles/0.5.0-CleanedMaster.csv")
```

```python
toDelete = ["Melting", "Boiling"]
cols = og.columns.tolist()
print(og.columns)

for col in cols:
    if "Melting" in col or "Boiling" in col:
        try: 
            og.drop(col, axis=1, inplace=True)
        except:
            pass

print(og.columns)
```

```python
logSPreds = pd.read_csv("../Results/logS/Prediction/LGBM/0.1.0_MRobust-2024-08-13_11.11.21_UnknownPreds.csv")
logHPreds = pd.read_csv("../Results/logHenry/Prediction/LGBM/0.1.0_MRobust-2024-08-13_11.11.21_UnknownPreds.csv")
```

```python
logSPreds.insert(2, "logS-dataSource", "LightGBM")
logSPreds = logSPreds.rename(columns = {"yUnk": "logS"})
logSPreds
```

```python
logHPreds.insert(2, "HenryConstant-dataSource", "LightGBM")
logHPredVals = logHPreds["yUnk"].values.tolist()
HPredVals = np.exp(logHPredVals)
logHPreds["HenryConstant"] = HPredVals
logHPreds
```

```python
df = og.copy()
targets = ["logS", "HenryConstant"]
df.dropna(subset = targets, inplace=True, how = "all")
df.drop_duplicates(inplace=True)
df
```

```python
masterInChI = df["InChI"].values.tolist()

hasHL = df.dropna(subset = ["HenryConstant"], how = "all")
hasHL = hasHL["InChI"].values.tolist()
print(len(masterInChI) - len(hasHL))

logHPredInChI = logHPreds["InChI"].values.tolist()
print(len(logHPredInChI))

intersect = list(set(hasHL) & set(logHPredInChI))
print(len(intersect))

for inchi in intersect: #Removing predictions that would overwrite true data
    logHPreds = logHPreds.drop(logHPreds[logHPreds["InChI"] == inchi].index)

logHPreds
```

```python
predHInChI = logHPreds["InChI"].values.tolist()

print(list(set(predHInChI) - (set(masterInChI) & set(predHInChI)))) #This should be 0. Checking for data that is not in the master dataframe.
```

```python
import warnings
warnings.filterwarnings("ignore")
from tqdm import tqdm
```

```python
copy = df.copy()
errors = []
print(copy.isna().sum())
for inchi in tqdm(predHInChI):
    subdf = copy[copy["InChI"] == inchi] 
    subpred = logHPreds[logHPreds["InChI"] == inchi]
    
    if subpred.shape[0] != subdf.shape[0]:
        print("Error with", inchi)
        errors.append(inchi)
    else:
        subdf["HenryConstant"] = subpred["HenryConstant"].values
        subdf["HenryConstant-dataSource"] = subpred["HenryConstant-dataSource"].values
    copy[copy["InChI"] == inchi] = subdf

print(copy.isna().sum())
```

```python
copy.to_csv("../Results/AugmentedHenrysLawData.csv", index=False)
```
