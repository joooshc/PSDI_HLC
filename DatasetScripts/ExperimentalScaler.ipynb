```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, normalize, PowerTransformer, MaxAbsScaler
from scipy import stats
from MiscScripts.DataCleaning import *
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../Data/Datasets/0.6.0-BPMP.csv")
```

```python
df.drop_duplicates(inplace=True)
mp = df.pop("MeltingPoint_C")
df.insert(0, "MeltingPoint_C", mp)
```

```python
df.iloc[:, 8:] #Select from col 8 onwards
```

```python
df = prepForScaling(df, "BoilingPoint_C", [])
df = prepForScaling(df, "MeltingPoint_C", [])

Norm = pd.DataFrame(normalize(df.iloc[:, 8:], axis=0), columns = df.iloc[:, 8:].columns)
Norm = dropLowDistinction(Norm, 0.001)
print("\n\nNormalized:", Norm.shape)

MinMax = pd.DataFrame(MinMaxScaler().fit_transform(df.iloc[:, 8:]), columns = df.iloc[:, 8:].columns)
MinMax = dropLowDistinction(MinMax, 0.001)
print("MinMax:", MinMax.shape)
```

```python
dsets = {"Norm": Norm, "MinMax": MinMax}
vers = "0.1.1"

methods = [StandardScaler(), RobustScaler(), PowerTransformer(standardize=False), PowerTransformer(standardize=True), MaxAbsScaler(), "None"]
```

```python
def logTarget(df, target):
    df.dropna(subset=[target], inplace=True)

    targVals = df[target].values.astype(float).tolist()
    targVals = targVals + abs(np.min(targVals)) + 1

    log = np.log1p(targVals)
    log = [x if x != -np.inf else np.nan for x in log]
    log = [x if x != np.inf else np.nan for x in log]
    df[target] = log

    df.dropna(subset=[target], inplace=True)
    return df
```

```python
for dkey in dsets.keys():
    dset = dsets[dkey]
    for method in methods:
        if method != "None":
            scaled = pd.DataFrame(method.fit_transform(dset), columns = dset.columns)
        else:
            scaled = dset
        scaled = pd.concat([df.iloc[:, :8], scaled], axis=1)
        scaled.dropna(subset=["BoilingPoint_C", "MeltingPoint_C"], inplace=True)
        print(dkey, method, scaled.shape)
        scaled.to_csv(f"../Data/Datasets/BPMP_Scaling/{vers}_{dkey}_{method}.csv", index=False)

        scaled = logTarget(scaled, "BoilingPoint_C")
        scaled = logTarget(scaled, "MeltingPoint_C")
        print("Log", dkey, method, scaled.shape)
        scaled.to_csv(f"../Data/Datasets/BPMP_Scaling/{vers}_{dkey}Log_{method}.csv", index=False)
```

```python
for dkey in dsets.keys():
    dset = dsets[dkey]
    scaled = pd.DataFrame(np.log1p(dset), columns = dset.columns)
    scaled = pd.concat([df.iloc[:, :8], scaled], axis=1)
    scaled.to_csv(f"../Data/Datasets/BPMP_Scaling/{vers}_{dkey}_Log.csv", index=False)

    scaled = logTarget(scaled, "BoilingPoint_C")
    scaled = logTarget(scaled, "MeltingPoint_C")
    scaled.to_csv(f"../Data/Datasets/BPMP_Scaling/{vers}_{dkey}Log_Log.csv", index=False)
```
