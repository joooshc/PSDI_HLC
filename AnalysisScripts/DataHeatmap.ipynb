```python
def createHeatmap(data, yLabels, title):
    fig, ax = plt.subplots(figsize=(25, 25), dpi=150, layout = "tight")
    ax = sns.heatmap(data, cmap = "viridis", yticklabels = yLabels, cbar = True, robust = "True", xticklabels=False)
    plt.title(title, fontsize = 40, weight = "bold")
    plt.show()
```

```python
def getData(df, toPop):
    popped = df.pop(toPop)
    data = df.iloc[:, 6:]
    # data.insert(0, toPop, popped)
    return data.T
```

```python
dfScaled = pd.read_csv("../Data/Datasets/PredictionDatasets/0.2.2-logHenry-MRobust-TrainSet.csv")
df = pd.read_csv("../Data/Datasets/0.7.1-hasLogH.csv")

scaled = getData(dfScaled, "logHenry")
unscaled = getData(df, "logHenry")

fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(30, 15), dpi=150, layout = "tight")

ax0 = sns.heatmap(unscaled, cmap = "viridis",  cbar = True, robust = "True", xticklabels=False, ax = ax0)
ax0.set_title("Heatmap of logH Training Set Features (Unscaled)", fontsize = 20, weight = "bold")

ax1 = sns.heatmap(scaled, cmap = "viridis", cbar = True, robust = "True", xticklabels=False, ax = ax1)
ax1.set_title("Heatmap of logH Training Set Features (Scaled by MinMax, then RobustScaler)", fontsize = 20, weight = "bold")
plt.suptitle("Comparison of the Feature Heatmaps for Scaled and Unscaled Data ($k_H$)", fontsize = 30, weight = "bold")
plt.show()
```
