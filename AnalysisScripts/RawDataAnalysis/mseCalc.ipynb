```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv("../Data/Datasets/Normalised/1.3.0-logHenry.csv")
df.drop_duplicates(subset="logHenry", inplace=True)
results = pd.read_csv("../Results/logHenry/LGBM/1.3.0_Y_2024-07-17_17.15.28.csv")

df.sort_values("logHenry", inplace=True)
df = df.iloc[:, 0:5]
print(df.shape)
```

```python
Y = results["Y"].values.reshape(-1, 1)
Y_pred = results["Y_hat"].values.reshape(-1, 1)
mse = [mean_squared_error(y, yi) for y, yi in zip(Y, Y_pred)]
mae = [mean_absolute_error(y, yi) for y, yi in zip(Y, Y_pred)]
results["MSE"] = mse
results["MAE"] = mae
results.sort_values("Y", inplace=True)
```

```python
dfCombined = pd.DataFrame()
cols = results.iloc[:, 1:4].columns.tolist()

for yval in df["logHenry"].values.tolist():
    subset1 = df[df["logHenry"] == yval]
    subset2 = results[results["Y"] == yval]

    for col in cols:
        try:
            insertVal = subset2[col].values[0]
        except:
            insertVal = np.nan
        
        subset1.insert(5, col, insertVal)

    dfCombined = pd.concat([dfCombined, subset1])
dfCombined.dropna(inplace=True)

dfCombined.sort_values("MSE", inplace=True, ascending=False)
dfCombined.reset_index(drop=True, inplace=True)
dfCombined
```
