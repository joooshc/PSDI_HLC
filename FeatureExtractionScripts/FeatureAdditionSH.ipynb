## Importing Libaries

```python
import pandas as pd
import numpy as np
from tqdm import tqdm 
import sys
sys.path.append('../')
from DatasetScripts.MiscScripts.dfToDict import *
```

## Importing Data

```python
path = "../Data/Combined/LargeFiles/" 
vers = "0.4.0"

data = pd.read_csv(f"../Data/Combined/0.4.1-CleanedMaster.csv")

rdkitFeatures = pd.read_csv(f"{path}ExtractedDescriptors/{vers}-RDKitDescriptors.csv")
thermoFeatures = pd.read_csv(f"{path}ExtractedDescriptors/{vers}-ThermoDesc.csv")
```

```python
print(rdkitFeatures.shape)
rdkitFeatures.drop_duplicates(inplace=True)
print(rdkitFeatures.shape)
print(thermoFeatures.shape)
thermoFeatures.drop_duplicates(inplace=True)
print(thermoFeatures.shape)

data.iloc[:, 0:13]
```

## Rearranging Columns

```python
# toInsert = ["MeltingPoint/C", "BoilingPoint/C", "logS", "HenryConstant", "Temperature"] #List of columns to move
toInsert = ["logS", "HenryConstant", "Temperature"] #List of columns to move

for col in toInsert: #Deleting and reinserting at beginning of dataframe
    x = data.pop(col)
    data.insert(0, col, x)
try:
    del data["SMILES"]
except:
    pass
data.iloc[:, 0:11] 
```

## Creating DataFrame of Constants

```python
thermoConstants = thermoFeatures.iloc[:, 0:30] #First 30 columns are constants
thermoDescriptors = thermoFeatures.iloc[:, 30:]

thermoDescriptors.insert(0, "InChI", thermoFeatures["InChI"]) #Adding InChI column to descriptors
thermoConstants.drop_duplicates(subset="InChI", inplace=True) #Dropping duplicates

rdkitFeatures.reset_index(drop=True, inplace=True) #Will not work without resetting index
thermoConstants.reset_index(drop=True, inplace=True)

rdkitFeatures.sort_values(by="InChI", inplace=True) #Sorting by InChI
thermoConstants.sort_values(by="InChI", inplace=True)
rdkitFeatures["InChI"] #Deleting duplicate column
rdkitFeatures.rename(columns={"MolWt": "MolWtRDKIt"}, inplace=True) #Renaming MolWt column

molwt = rdkitFeatures["MolWtRDKIt"].values.tolist()
consMolWt = thermoConstants["MolWt"].values.tolist()
```

```python
try:
    constants = pd.read_csv(f"{path}ExtractedDescriptors/{vers}-RDKitThermoConsts.csv")

except:
    inchiList = data["InChI"].tolist() #List of InChI strings
    inchiList = np.unique(inchiList) #Removing duplicates

    constants = pd.DataFrame()
    for inchi in tqdm(inchiList, desc="Creating constants dataframe"):
        subThermo = thermoConstants[thermoConstants["InChI"] == inchi]
        subRDKit = rdkitFeatures[rdkitFeatures["InChI"] == inchi]

        subThermoDict = dfToDict(subThermo)
        subRDKitDict = dfToDict(subRDKit)

        subThermoDict = subThermoDict | subRDKitDict
        df = pd.DataFrame(subThermoDict)

        constants = pd.concat([constants, df])

    constants = constants.fillna(value=np.nan)
    constants.reset_index(drop=True, inplace=True)
    constants.to_csv(f"{path}ExtractedDescriptors/{vers}-RDKitThermoConsts.csv", index=False)
```

## Adding to Main Dataset

```python
inchiList = data["InChI"].values
inchiList = np.unique(inchiList) #Creating list of unique inchi values

master = pd.DataFrame() #Initialising dataframe

for inchi in tqdm(inchiList):
    subData = data[data["InChI"] == inchi] #Selecting data and resetting indexes
    # subData = subData.iloc[:, 0:13]
    subData.reset_index(drop=True, inplace=True)

    subFeatures = thermoDescriptors[thermoDescriptors["InChI"] == inchi]
    subFeatures.reset_index(drop=True, inplace=True)

    subConstants = constants[constants["InChI"] == inchi]
    subConstants.reset_index(drop=True, inplace=True)

    df = pd.concat([subData, subFeatures], ignore_index=True) #Concatenating non constants
    subConstants = pd.concat([subConstants]*df.shape[0], ignore_index=True) #Repeating
    del subConstants["InChI"] #Deleting duplicate column

    df = pd.concat([subData, subConstants], axis=1) #Concatenating the two dataframes
    master = pd.concat([master, df], axis=0) #Concatenating to master dataframe

toInsert = ["logS-dataSource","HenryConstant-dataSource", "nIsomers", "Temperature", "InChI"] #Rearranging
for col in toInsert:
    x = master.pop(col)
    master.insert(0, col, x)

print("Intial shape: ", master.shape) #Dropping duplicates and NaNs
master.drop_duplicates(inplace=True)
master.dropna(subset=["InChI"], inplace=True)
print("Final shape: ", master.shape)

master
master.reset_index(drop=True, inplace=True)
master.to_csv(f"../Data/Combined/0.4.1-Master+Consts.csv", index=False)
```
